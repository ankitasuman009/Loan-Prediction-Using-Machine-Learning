<h2 style="color:black">Bi-variant plotting of attributes with target variable</h2>

```python
Gender = pd.crosstab(train1['Gender'], train1['Loan_Status'])
Gender.div(Gender.sum(1).astype(float), axis=0).plot(kind="bar", stacked=True, figsize=(4,4))
plt.show()
```

```python
Married = pd.crosstab(train1['Married'], train1['Loan_Status'])
Married.div(Married.sum(1).astype(float), axis=0).plot(kind="bar", stacked=True, figsize=(4,4))
plt.show()

Self_Employed = pd.crosstab(train1['Self_Employed'], train1['Loan_Status'])
Self_Employed.div(Self_Employed.sum(1).astype(float), axis=0).plot(kind="bar", stacked=True, figsize=(4,4))
plt.show()

Credit_History = pd.crosstab(train1['Credit_History'], train1['Loan_Status'])
Credit_History.div(Credit_History.sum(1).astype(float), axis=0).plot(kind="bar", stacked=True, figsize=(4,4))
plt.show()
```

```python
Dependents = pd.crosstab(train1['Dependents'], train1['Loan_Status'])
Dependents.div(Dependents.sum(1).astype(float), axis=0).plot(kind="bar", stacked=True, figsize=(4,4))
plt.show()
Education = pd.crosstab(train1['Education'], train1['Loan_Status'])
Education.div(Education.sum(1).astype(float), axis=0).plot(kind="bar", stacked=True, figsize=(4,4))
plt.show()
Property_Area =pd.crosstab(train1['Property_Area'], train1['Loan_Status'])
Property_Area.div(Property_Area.sum(1).astype(float), axis=0).plot(kind="bar", stacked=True, figsize=(4,4))
plt.show()
```

```python
train1.groupby("Loan_Status")["ApplicantIncome"].mean().plot.bar()
```

```python
bins = [0, 2500, 4000, 6000, 81000]
group = ['Low', 'Average', 'High', 'Very High']
train1['Income_bin'] = pd.cut(train1['ApplicantIncome'], bins, labels = group)
train1.head()
```

```python
Income_bin = pd.crosstab(train1['Income_bin'], train1['Loan_Status'])
Income_bin.div(Income_bin.sum(1).astype(float), axis=0).plot(kind="bar", stacked=True)
plt.xlabel("ApplicantIncome")
plt.ylabel("Percentage")
```

```python
bins = [0, 1000, 3000, 42000]
group = ['Low', 'Average', 'High']
train1['Coapplicant_Income_bin'] = pd.cut(train1['CoapplicantIncome'], bins, labels=group)
Coapplicant_Income_bin = pd.crosstab(train1['Coapplicant_Income_bin'], train1['Loan_Status'])
Coapplicant_Income_bin.div(Coapplicant_Income_bin.sum(1).astype(float), axis=0).plot(kind="bar", stacked=True)
plt.xlabel("CoapplicantIncome")
plt.ylabel("Percentage")
```

```python
train1['Total_Income'] = train1['ApplicantIncome'] + train1['CoapplicantIncome']
```

```python
bins = [0, 2500, 4000, 6000, 81000]
group = ['Low', 'Average', 'High', 'Very High']
train1['Total_Income_bin'] = pd.cut(train1['Total_Income'], bins, labels = group)
Total_Income_bin = pd.crosstab(train1['Total_Income_bin'], train1['Loan_Status'])
Total_Income_bin.div(Total_Income_bin.sum(1).astype(float), axis=0).plot(kind="bar", stacked=True)
plt.xlabel("Total_Income")
plt.ylabel("Percentage")
```

```python
bins = [0, 100, 200, 700]
group = ['Low', 'Average', 'High']
train1['Loan_Amount_bin'] = pd.cut(train1["LoanAmount"], bins, labels=group)
Loan_Amount_bin = pd.crosstab(train1['Loan_Amount_bin'], train1['Loan_Status'])
Loan_Amount_bin.div(Loan_Amount_bin.sum(1).astype(float), axis=0).plot(kind="bar", stacked=True)
plt.xlabel("Loan_Amount_bin")
plt.ylabel("Percentage")
```

```python
train2 = train1.drop(["Income_bin", 'Coapplicant_Income_bin', 'Total_Income', 'Total_Income_bin', 'Loan_Amount_bin'], axis=1)
train2.head()
```

```python
train2['Dependents'].replace('3+', 3, inplace=True)
test1['Dependents'].replace('3+', 3, inplace=True)
train2['Loan_Status'].replace('N', 0, inplace=True)
train2['Loan_Status'].replace('Y', 1, inplace=True)
```

```python
matrix = train2.corr()
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(matrix, vmax=0.8, square=True, cmap="BuPu", annot=True)
```
